```python
# ================================================
# Dimensionality Reduction using LLE and t-SNE
# Author: Finson N
# ================================================

# Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import LocallyLinearEmbedding, TSNE

# Step 2: Load and Preprocess Dataset
digits = load_digits()
X = digits.data
y = digits.target

# Standardize features
X_scaled = StandardScaler().fit_transform(X)

print("Data shape:", X_scaled.shape)

# Step 3: Apply Locally Linear Embedding (LLE)
lle = LocallyLinearEmbedding(n_neighbors=30, n_components=2, method='standard')
X_lle = lle.fit_transform(X_scaled)

# Step 4: Apply t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# Step 5: Visualize Results
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# LLE Visualization
sc1 = ax[0].scatter(X_lle[:, 0], X_lle[:, 1], c=y, cmap='tab10', s=15)
ax[0].set_title("Locally Linear Embedding (LLE)")
ax[0].set_xlabel("Component 1")
ax[0].set_ylabel("Component 2")
plt.colorbar(sc1, ax=ax[0])

# t-SNE Visualization
sc2 = ax[1].scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10', s=15)
ax[1].set_title("t-SNE")
ax[1].set_xlabel("Component 1")
ax[1].set_ylabel("Component 2")
plt.colorbar(sc2, ax=ax[1])

plt.tight_layout()
plt.show()
```
